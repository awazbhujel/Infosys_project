**Streamlit**

 is an open-source Python library that allows developers to build and deploy interactive web applications for data visualization and machine learning models easily. It is particularly popular among data scientists and machine learning practitioners because of its simplicity and focus on displaying data interactively with minimal effort.

**Features of Streamlit**

1. Ease of Use: Write an app using pure Python scripts; no need for complex frontend frameworks.
2. Interactive Widgets: Built-in widgets like sliders, dropdowns, and text inputs allow for interactivity with minimal code.
3. Live Updates: Automatically updates the web application as you modify the script.
4. Integration with Libraries: Seamlessly integrates with libraries like Pandas, Matplotlib, Plotly, and others.
5. Deployment: Easily deploy applications using platforms like Streamlit Community Cloud or other cloud platforms.
6. Customizable UI: Provides options to customize the layout and widgets to suit your application’s needs.
7. Data Visualization: Supports rendering of rich media like charts, maps, and dataframes.
8. Real-Time Applications: Useful for real-time applications like dashboards and ML model inference.

1. Declarative Programming
Streamlit uses a declarative approach, where the flow of the app is determined by the sequence of commands in the script. Every time the app runs, it executes the code from top to bottom, reflecting any changes immediately.

Example:

```python
import streamlit as st

st.title("Declarative Example")
st.write("Streamlit executes this script from top to bottom.")
```

2. Widgets
Streamlit provides a wide variety of interactive widgets to enable user interaction. These widgets can hold values that dynamically control your application.

Common Widgets:

Slider: Allows users to select a value in a range.

File Uploader: Upload files like CSVs or images.

Text Input: Accept textual input from users.

Button: Trigger specific actions.

```python
age = st.slider("Select your age", 1, 100, 25)
st.write(f"You selected age: {age}")
```

Sidebar
The sidebar is a secondary section where you can place widgets and other components.

Example:

```python
st.sidebar.title("Sidebar Example")
st.sidebar.slider("Adjust a value", 0, 50, 25)
```

Streamlit Commands Summary

st.selectbox()	Creates a dropdown selection box.

st.plotly_chart()	Displays a Plotly chart.

st.session_state	Stores values across interactions.
