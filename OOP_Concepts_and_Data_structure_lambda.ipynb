**Object-Oriented Programming (OOP) is a programming paradigm that organizes and models data and functionality into objects and classes. Python supports OOP concepts, which makes it a versatile and popular language for various applications.**

**Why Use OOPS?**

1. Organized Code: Keeps related data and actions together.
2. Reuse: Use existing blueprints (classes) to create new objects or classes.
3. Easy Maintenance: Fixing and updating code is simpler.
4. Real-World Modeling: Makes programming closer to real-life scenarios.


**OOP Concepts in Python:**

**1. Class**

A class is a blueprint for creating objects. It defines the structure and behavior of its objects, including attributes (data) and methods (functions).

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand  # Attribute
        self.model = model  # Attribute

    def display_info(self):  # Method
        print(f"Car: {self.brand} {self.model}")
```

**2. Object**

An object is an instance of a class. It is created using the class constructor and can access class attributes and methods.

```python
my_car = Car("Toyota", "Corolla")
my_car.display_info()  # Output: Car: Toyota Corolla
```

3. **Encapsulation**

Encapsulation is the concept of bundling data (attributes) and methods (functions) within a class and restricting access to certain details to protect the object's integrity.

Private Attributes/Methods: Use a double underscore __ to make attributes or methods private.
Getters and Setters: Provide controlled access to private data.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private attribute

    def get_balance(self):  # Getter
        return self.__balance

    def deposit(self, amount):  # Setter
        if amount > 0:
            self.__balance += amount

account = BankAccount(1000)
print(account.get_balance())  # Output: 1000
account.deposit(500)
print(account.get_balance())  # Output: 1500
```

4. **Inheritance**

Inheritance allows a class (child) to inherit attributes and methods from another class (parent). This promotes code reuse and a hierarchical class structure.

Single Inheritance: A child class inherits from one parent class.
Multiple Inheritance: A child class inherits from multiple parent classes.

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def start(self):
        print(f"{self.brand} is starting.")

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)  # Call parent class constructor
        self.model = model

    def display_info(self):
        print(f"Car: {self.brand} {self.model}")

my_car = Car("Honda", "Civic")
my_car.start()  # Output: Honda is starting.
my_car.display_info()  # Output: Car: Honda Civic
```

**5. Polymorphism**

Polymorphism allows objects to take on multiple forms. It refers to using a single interface to represent different types or behaviors.

Method Overriding: A child class redefines a method from the parent class.
Duck Typing: Python's dynamic nature allows different objects to use the same method if it exists in their class.

```python
class Animal:
    def sound(self):
        print("Some generic sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")

animals = [Dog(), Cat()]
for animal in animals:
    animal.sound()
# Output:
# Bark
# Meow
```

**6. Abstraction**

Abstraction hides the complexity of implementation and only exposes essential features. It is typically achieved using abstract classes and methods.

Abstract Class: Defined using the abc module and contains abstract methods that must be implemented in child classes.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

rect = Rectangle(10, 20)
print(rect.area())  # Output: 200
```

**Opps concepts example**

```python
from abc import ABC, abstractmethod  # For abstraction

# Abstract class to demonstrate abstraction
class Vehicle(ABC):
    def __init__(self, brand, color):
        self.brand = brand  # Public attribute
        self._color = color  # Protected attribute

    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass

    def display_info(self):
        print(f"Vehicle: {self.brand}, Color: {self._color}")


# Subclass inheriting from Vehicle
class Car(Vehicle):
    def __init__(self, brand, color, model):
        super().__init__(brand, color)
        self.__model = model  # Private attribute

    # Overriding the abstract method
    def start(self):
        print(f"{self.brand} {self.__model} is starting.")

    def stop(self):
        print(f"{self.brand} {self.__model} is stopping.")

    def display_info(self):
        super().display_info()  # Call parent class method
        print(f"Model: {self.__model}")


# Another subclass inheriting from Vehicle
class Bike(Vehicle):
    def __init__(self, brand, color, engine_capacity):
        super().__init__(brand, color)
        self.engine_capacity = engine_capacity  # Public attribute

    # Overriding the abstract method
    def start(self):
        print(f"{self.brand} bike with {self.engine_capacity}cc engine is starting.")

    def stop(self):
        print(f"{self.brand} bike with {self.engine_capacity}cc engine is stopping.")


# Polymorphism demonstration
def test_vehicle(vehicle):
    vehicle.start()
    vehicle.display_info()
    vehicle.stop()


# Main program
if __name__ == "__main__":
    # Create objects for Car and Bike
    car = Car("Toyota", "Red", "Corolla")
    bike = Bike("Yamaha", "Blue", 150)

    # Test car and bike using polymorphism
    print("Testing Car:")
    test_vehicle(car)

    print("\nTesting Bike:")
    test_vehicle(bike)
```

#**Data structure in Python**

**Difference Between List, Tuple, Set, and Dictionary**

1. List:
A list is an ordered collection of elements that can contain duplicates. You can modify its contents (add, remove, or update elements). Lists maintain the order of insertion, and you can access items using indexing.

2. Tuple:
A tuple is similar to a list but is immutable, meaning its contents cannot be changed once created. It also maintains order and allows duplicates. Tuples are used when you want to ensure data remains constant.

3. Set:
A set is an unordered collection of unique elements. It does not allow duplicates and does not maintain any specific order of elements. Sets are ideal for operations like union, intersection, and difference.

4. Dictionary:
A dictionary is an unordered collection of key-value pairs. Keys are unique and must be immutable, while values can be of any type and can have duplicates. Dictionaries are used to map keys to corresponding values for fast lookups.

1. **List**

Ordered: Maintains the order of elements.

Mutable: Can modify elements (add, remove, or change values).

Duplicates: Allows duplicate elements.

Indexing/Slicing: Supports indexing and slicing.

Syntax: Defined using square brackets [].

```python
lst = [1, 2, 3]
lst.append(4)          # [1, 2, 3, 4]
lst.extend([5, 6])      # [1, 2, 3, 4, 5, 6]
lst.insert(2, 99)       # [1, 2, 99, 3, 4, 5, 6]
lst.remove(99)          # [1, 2, 3, 4, 5, 6]
print(lst.pop())        # Output: 6; List: [1, 2, 3, 4, 5]
lst.reverse()           # [5, 4, 3, 2, 1]
print(lst)
```

**2. Tuple**

Ordered: Maintains the order of elements.

Immutable: Cannot modify elements after creation.

Duplicates: Allows duplicate elements.

Indexing/Slicing: Supports indexing and slicing.

Syntax: Defined using parentheses ().

```python
tup = (1, 2, 3, 2, 4, 2)
print(tup.count(2))      # Output: 3
print(tup.index(3))      # Output: 2

# Indirect Modification: Convert to List
tup_list = list(tup)
tup_list.append(5)       # [1, 2, 3, 2, 4, 2, 5]
tup = tuple(tup_list)    # Back to tuple
print(tup)               # (1, 2, 3, 2, 4, 2, 5)
```

**3. Set**

Unordered: Does not maintain the order of elements.

Mutable: Can add or remove elements, but individual elements are immutable.

No Duplicates: Does not allow duplicate elements.

Indexing/Slicing: Not supported since it’s unordered.

Syntax: Defined using curly braces {} or the set() function.

```python
s = {1, 2, 3}
s.add(4)                # {1, 2, 3, 4}
s.remove(2)             # {1, 3, 4}
s.discard(5)            # No error, still {1, 3, 4}
print(s.pop())          # Removes an arbitrary element
s.update([5, 6])        # {3, 4, 5, 6}
s.clear()               # Empty set: {}
print(s)
```

**4. Dictionary**

Ordered (Python 3.7+): Maintains the order of key-value pairs.

Mutable: Can modify keys or values.

No Duplicate Keys: Keys must be unique; values can be duplicated.

Key-Value Pairs: Stores data as key-value pairs.

Indexing: Accessed via keys, not positions.

Syntax: Defined using curly braces {} with key-value pairs.

```python
d = {'name': 'Alice', 'age': 25}
print(d.get('name'))         # Output: Alice
d.update({'city': 'NYC'})    # {'name': 'Alice', 'age': 25, 'city': 'NYC'}
print(d.pop('age'))          # Output: 25; Dict: {'name': 'Alice', 'city': 'NYC'}
print(d.popitem())           # Output: ('city', 'NYC'); Dict: {'name': 'Alice'}
d['country'] = 'USA'         # {'name': 'Alice', 'country': 'USA'}
print(d.keys())              # dict_keys(['name', 'country'])
print(d.values())            # dict_values(['Alice', 'USA'])
print(d.items())             # dict_items([('name', 'Alice'), ('country', 'USA')])
d.clear()                    # Empty dictionary: {}
print(d)
```

**Lambda function**

A lambda function in Python is a small, anonymous function that can have any number of arguments but only one expression. It is defined using the lambda keyword and is often used for short-term operations, especially in cases where defining a full function might be overkill.

**Why Lambda Functions Are Used?**

Lambda functions are used when you need a small, simple function for a short period, especially when defining a full function would be unnecessarily verbose. They are particularly useful in scenarios involving functional programming and inline operations.

```python
# Regular function
def add(x, y):
    return x + y

# Lambda function
add = lambda x, y: x + y
print(add(10,20))
```

```python
nums = [1, 2, 3, 4]
squares = map(lambda x: x**2, nums)
print(list(squares))  # Output: [1, 4, 9, 16]
```
